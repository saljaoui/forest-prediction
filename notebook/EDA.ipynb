```python
import pandas as pd
import numpy as np

test = pd.read_csv('../data/test.csv')
train = pd.read_csv('../data/train.csv')
from matplotlib import pyplot as plt
def eda(df):
    print(df.shape, "\n")

    print("number of missing values",df.isnull().sum().sum(), "\n")

    print("number of duplicated values",df.duplicated().sum(), "\n")

    non_int_cols = df.select_dtypes(exclude=['int64']).columns

    if len(non_int_cols) == 0:
        print("ALL columns are int64")
    else:
        print("NOT Columns are int64:")
    print("Data types:\n", df.dtypes, "\n")
    print("Statistical Summary:\n", df.describe(include='all'), "\n")
    for col in df.columns:
        print(f"Column: {col}")
        print(f"Unique values: {df[col].nunique()}")
        print(f"Missing values: {df[col].isnull().sum()}")
        print("\n")
```

```python
eda(test)
```

```python
eda(train)
           
```
